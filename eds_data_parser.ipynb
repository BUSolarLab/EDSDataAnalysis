```python
# construct the weeks
#list(df['Date'])
day_range = 7

a = list(df['OCV(V)'])
counter = 0
pre = []
post = []
for i in range(len(a)):
    if counter % 2 == 0:
        pre.append(a[i])
        counter = counter + 1
    else:
        post.append(a[i])
        counter = counter + 1

        
```
